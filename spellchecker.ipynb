```python
import re
def correct_spelling_and_count(text, custom_words):
    spell = SpellChecker()
    spell.word_frequency.load_words(custom_words)  # Load custom words

    # Split the text into sentences
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)

    corrected_sentences = []
    corrected_count = 0
    uncorrected_count = 0

    for sentence in sentences:
        # Split the sentence into words while keeping punctuation intact
        words = re.findall(r"\b\w+\b", sentence)
        corrected_words = []
        sentence_corrected = False

        for word in words:
            # If the word is already correct, leave it as is
            # Use spell.known([word]) instead of spell.words
            if word in spell.known([word]):
                corrected_words.append(word)
            else:
                # Check suggestions and replace with the most likely correction
                corrected_word = spell.correction(word)
                # Ensure the correction exists (can handle None)
                corrected_word = corrected_word if corrected_word else word
                corrected_words.append(corrected_word)
                if corrected_word != word:
                    sentence_corrected = True

        # Reconstruct the sentence from corrected words
        corrected_sentence = sentence
        for original, corrected in zip(words, corrected_words):
            corrected_sentence = re.sub(rf"\b{original}\b", corrected, corrected_sentence, count=1)

        corrected_sentences.append(corrected_sentence)

        # Count corrected and uncorrected sentences
        if sentence_corrected:
            corrected_count += 1
        else:
            uncorrected_count += 1

    # Rejoin sentences into corrected text
    corrected_text = " ".join(corrected_sentences)

    return corrected_text, corrected_count, uncorrected_count

```
